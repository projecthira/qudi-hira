```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sample_rate = 1.25e9

null_pulse_length = 10e-9
delay_length = 1e-6
pi_pulse_length = 200e-9
laser_readout_length = 300e-9

freq_list = np.arange(0, 100)
freq_rep = 100

total_pulse_length = null_pulse_length + delay_length + pi_pulse_length + laser_readout_length
total_pulse_samples = int(total_pulse_length * sample_rate * freq_rep * len(freq_list))

delay_pulse_sample = np.zeros(int(np.floor(sample_rate * delay_length)))
switch_pulse_sample = np.zeros(int(np.floor(sample_rate * pi_pulse_length)))
null_pulse_sample = np.zeros(int(np.floor(null_pulse_length * sample_rate)))
laser_readout_pulse_sample = np.ones(int(np.floor(sample_rate * laser_readout_length)))


laser_sequence = np.zeros(total_pulse_samples * 2)
laser_sequence[:] = np.NaN
```

```python
idxa = 0
for idx_freq_list, f in enumerate(freq_list):    
    idxb = idxa + len(null_pulse_sample)
    laser_sequence[idxa:idxb] = np.zeros(idxb-idxa)
    print("-------")

    print("a", idxa)
    
    print("b", idxb)
    
    idx4 = idxb
    for idx_freq_rep in range(freq_rep):
        idx1 = idx4
        print("1", idx1)
        
        idx2 = idx1 + len(null_pulse_sample)
        laser_sequence[idx1:idx2] = np.zeros(idx2-idx1)
        print("2", idx2)

        idx3 = idx2 + len(laser_readout_pulse_sample)
        laser_sequence[idx2:idx3] = np.ones(idx3-idx2)
        print("3", idx3)
        
        idx4 = idx3 + len(delay_pulse_sample) + len(switch_pulse_sample) + len(null_pulse_sample)
        print("4", idx4)
        laser_sequence[idx3:idx4] = np.zeros(idx4-idx3)
        
        print("******")
        
    idxa = idx4
    
laser_sequence = laser_sequence[~np.isnan(laser_sequence)]
```
